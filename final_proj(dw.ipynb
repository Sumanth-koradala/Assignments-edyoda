```python
data.dropna(subset = ['continent'],inplace = True)
data
```

```python
data.fillna(0,inplace = True)
data
```

```python
data['date'] = pd.to_datetime(data['date'])
data
```

```python
data['month'] = pd.DatetimeIndex(data['date']).month
data
```

```python
data.groupby('continent').max().reset_index()
```

```python
df_groupby = data.groupby('continent').max().reset_index()

print(df_groupby)
```

```python
df_groupby = data['total_deaths_to_total_cases'] = data['total_deaths'] / data['total_cases']
df_groupby
```

```python
df_groupby = sns.displot(data['gdp_per_capita'])
df_groupby
```

```python
sns.scatterplot(data=data, x='total_cases', y='gdp_per_capita')

plt.xlabel('Total Cases')
plt.ylabel('GDP per Capita')
plt.title('Scatter Plot of Total Cases vs GDP per Capita')

print(plt.show())
```

```python
df_groupby = data.groupby(["continent"]).agg({"total_cases":"sum", "total_deaths":"sum", "population":"mean"}).reset_index()
sns.pairplot(df_groupby)
```

```python
df_groupby = sns.catplot(data= data, kind='bar', x='continent', y='total_cases')
df_groupby
```
